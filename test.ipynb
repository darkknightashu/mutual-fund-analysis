```python
import pandas as pd
import os

# folder path
data_path = r"D:\test\data"

# load csv files
fund_master = pd.read_csv(os.path.join(data_path, "01_fund_master.csv"))
nav_history = pd.read_csv(os.path.join(data_path, "02_nav_history.csv"))
aum = pd.read_csv(os.path.join(data_path, "03_aum_by_fund_house.csv"))
sip = pd.read_csv(os.path.join(data_path, "04_monthly_sip_inflows.csv"))
category = pd.read_csv(os.path.join(data_path, "05_category_inflows.csv"))
folio = pd.read_csv(os.path.join(data_path, "06_industry_folio_count.csv"))
performance = pd.read_csv(os.path.join(data_path, "07_scheme_performance.csv"))
transactions = pd.read_csv(os.path.join(data_path, "08_investor_transactions.csv"))
holdings = pd.read_csv(os.path.join(data_path, "09_portfolio_holdings.csv"))
benchmark = pd.read_csv(os.path.join(data_path, "10_benchmark_indices.csv"))

print("All files loaded successfully")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

sip["month"] = pd.to_datetime(sip["month"])

plt.figure(figsize=(10,5))
plt.plot(sip["month"], sip["sip_inflow_cr"])

plt.title("Monthly SIP Inflows Trend")
plt.xlabel("Month")
plt.ylabel("Crores")
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

sip["month"] = pd.to_datetime(sip["month"])

plt.figure(figsize=(10,5))
plt.plot(sip["month"], sip["sip_inflow_crore"])

plt.title("Monthly SIP Inflows Trend")
plt.xlabel("Month")
plt.ylabel("Crores")
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

top_aum = (
    aum.groupby("fund_house")["aum_crore"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

plt.figure(figsize=(10,5))
top_aum.plot(kind="bar")

plt.title("Top 10 Fund Houses by AUM")
plt.xlabel("Fund House")
plt.ylabel("AUM (Crore)")
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

top_categories = (
    category.groupby("category")["net_inflow_crore"]
    .sum()
    .sort_values(ascending=False)
)

plt.figure(figsize=(10,5))
top_categories.plot(kind="bar")

plt.title("Net Inflows by Category")
plt.xlabel("Category")
plt.ylabel("Net Inflow (Crore)")
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

top_perf = (
    performance.sort_values("returns_1yr_pct", ascending=False)
    .head(10)
)

plt.figure(figsize=(10,5))
plt.bar(top_perf["scheme_name"], top_perf["returns_1yr_pct"])

plt.title("Top 10 Schemes by 1-Year Return")
plt.xlabel("Scheme")
plt.ylabel("1-Year Return %")

plt.xticks(rotation=90)

plt.tight_layout()
plt.show()
```
